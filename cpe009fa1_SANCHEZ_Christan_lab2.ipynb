Creating Character Class

```python
class Character:
    def __init__(self, username):
        self.username = username
        self.level = 1
        self.hp = 100
        self.dmg = 5
        self.str = 0
        self.agi = 0
        self.vit = 0
        self.int = 0

    def getUsername(self):
        return self.username

    def getLevel(self):
        return self.level

    def getHp(self):
        return self.hp

    def getDmg(self):
        return self.dmg

    def getStr(self):
        return self.str

    def getAgi(self):
        return self.agi

    def getVit(self):
        return self.vit

    def getInt(self):
        return self.int

    def setStr(self, new_str):
        self.str = new_str

    def setAgi(self, new_agi):
        self.agi = new_agi

    def setVit(self, new_vit):
        self.vit = new_vit

    def setInt(self, new_int):
        self.int = new_int

    def setDmg(self, new_dmg):
        self.dmg = new_dmg

    def setHp(self, new_hp):
        self.hp = new_hp

    def reduceHp(self, amount):
        self.hp -= amount

    def addHp(self, amount):
        self.hp += amount

character1 = Character("Elesis")
print(character1.getUsername()) # Access the username using the public method
```

Single Level Inheritance

```python
class Novice(Character):
    def basicAttack(self, character):
        character.reduceHp(self.getDmg())  # Reduces HP of target by damage
        print(f"{self.getUsername()} performed Basic Attack! - {self.getDmg()} damage")


character1 = Novice("Elesis") # Creates a Novice object
print(character1.getUsername()) # Prints the username of the created object
print(character1.getHp()) # Prints the HP which was inherited from the gethp in the character class
```

Multi-level Inheritance


```python

class Swordsman(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setStr(5)  # Fixed setStr here
        self.setDmg(10)
        self.setHp(self.getHp() + self.getVit())

    def slashAttack(self, character):
        new_dmg = self.getDmg() + self.getStr()
        character.reduceHp(new_dmg)
        print(f"{self.getUsername()} performed Slash Attack! - {new_dmg} damage")

class Archer(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setAgi(5)
        self.setInt(5)
        self.setVit(5)
        self.setHp(self.getHp() + self.getVit())

    def rangedAttack(self, character):
        new_dmg = self.getDmg() + self.getAgi()
        character.reduceHp(new_dmg)
        print(f"{self.getUsername()} performed Ranged Attack! - {new_dmg} damage")

class Magician(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setInt(10)
        self.setVit(5)
        self.setHp(self.getHp() + self.getVit())

    def heal(self):
        self.addHp(self.getInt())
        print(f"{self.getUsername()} performed Heal! - {self.getInt()} HP restored")

    def fireball(self, character):
        new_dmg = self.getDmg() + self.getInt()
        character.reduceHp(new_dmg)
        print(f"{self.getUsername()} performed Fireball Attack! - {new_dmg} damage")

class Boss(Swordsman, Archer, Magician):
    def __init__(self, username):
        super().__init__(username)
        self.setStr(10)
        self.setVit(25)
        self.setInt(5)
        self.setHp(self.getHp() + self.getVit())

    def slashAttack(self, character):
        Swordsman.slashAttack(self,character)  # Uses the Swordsman method

    def rangedAttack(self, character):
        Archer.rangedAttack(self,character)  # Uses the Archer method

    def fireball(self, character):
        Magician.fireball(self,character)  # Uses the Magician method
```

```python


# Creating Swordsman and Boss objects
character1 = Swordsman("Laxus")
character2 = Boss("Thanatos")

# Print initial HP
print(f"{character1.getUsername()} HP: {character1.getHp()}")
print(f"{character2.getUsername()} HP: {character2.getHp()}")

# Perform attacks
character1.slashAttack(character2)
print(f"{character2.getUsername()} HP: {character2.getHp()}")
character2.heal()  # Boss heals
character2.slashAttack(character1)  # Boss performs slash attack
character2.rangedAttack(character1)  # Boss performs ranged attack
character2.fireball(character1)  # Boss performs fireball attack

# Print final HP
print(f"{character1.getUsername()} HP: {character1.getHp()}")
print(f"{character2.getUsername()} HP: {character2.getHp()}")

```

```python
import random

class Character:
    def __init__(self, username):
        self.__username = username
        self.__hp = 100
        self.__mana = 100
        self.__dmg = 5
        self.__str = 0  # Strength stat
        self.__vit = 0  # Vitality stat
        self.__int = 0  # Intelligence stat
        self.__agi = 0  # Agility stat

    def getUsername(self):
        return self.__username

    def getHp(self):
        return self.__hp

    def setHp(self, new_hp):
        self.__hp = new_hp

    def getDmg(self):
        return self.__dmg

    def setDmg(self, new_dmg):
        self.__dmg = new_dmg

    def getStr(self):
        return self.__str

    def setStr(self, new_str):
        self.__str = new_str

    def getVit(self):
        return self.__vit

    def setVit(self, new_vit):
        self.__vit = new_vit

    def getInt(self):
        return self.__int

    def setInt(self, new_int):
        self.__int = new_int

    def getAgi(self):
        return self.__agi

    def setAgi(self, new_agi):
        self.__agi = new_agi

    def reduceHp(self, dmg_amt):
        self.__hp -= dmg_amt

    def addHp(self, heal_amt):
        self.__hp += heal_amt


class Novice(Character):
    def basicAttack(self, character):
        character.reduceHp(self.getDmg())
        print(f"{self.getUsername()} performed Basic Attack! - {self.getDmg()} damage")


class Swordsman(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setStr(5)
        self.setDmg(10)
        self.setHp(self.getHp() + self.getVit())

    def slashAttack(self, character):
        new_dmg = self.getDmg() + self.getStr()
        character.reduceHp(new_dmg)
        print(f"{self.getUsername()} performed Slash Attack! - {new_dmg} damage")


class Archer(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setAgi(5)
        self.setInt(5)
        self.setVit(5)
        self.setHp(self.getHp() + self.getVit())

    def rangedAttack(self, character):
        new_dmg = self.getDmg() + self.getAgi()
        character.reduceHp(new_dmg)
        print(f"{self.getUsername()} performed Ranged Attack! - {new_dmg} damage")


class Magician(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setInt(10)
        self.setVit(5)
        self.setHp(self.getHp() + self.getVit())

    def heal(self):
        self.addHp(self.getInt())
        print(f"{self.getUsername()} performed Heal! - {self.getInt()} HP restored")

    def fireball(self, character):
        new_dmg = self.getDmg() + self.getInt()
        character.reduceHp(new_dmg)
        print(f"{self.getUsername()} performed Fireball Attack! - {new_dmg} damage")


class Boss(Character):
    def __init__(self, username):
        super().__init__(username)
        self.setStr(10)
        self.setVit(25)
        self.setInt(5)
        self.setHp(self.getHp() + self.getVit())


class Game:
    def __init__(self):
        self.matches = []

    def start_game(self):
        mode = input("Select game mode (1 for Single Player, 2 for Player vs Player): ")
        if mode == '1':
            self.single_player_mode()
        elif mode == '2':
            self.player_vs_player_mode()
        else:
            print("Invalid selection. Please choose 1 or 2.")

    def single_player_mode(self):
        username = input("Enter your username: ")
        player = Novice(username)
        boss = Boss("Monster")
        wins = 0

        while player.getHp() > 0 and boss.getHp() > 0:
            if random.choice([True, False]):
                player.basicAttack(boss)
            else:
                boss.reduceHp(10)
                print(f"{boss.getUsername()} dealt 10 damage to {player.getUsername()}!")

            print(f"{player.getUsername()} HP: {player.getHp()} | {boss.getUsername()} HP: {boss.getHp()}")
            if boss.getHp() <= 0:
                print(f"{player.getUsername()} defeated the boss!")
                wins += 1
                if wins >= 2:
                    role = input("Choose your new role (Swordsman, Archer, Magician): ")
                    if role.lower() == 'swordsman':
                        player = Swordsman(username)
                    elif role.lower() == 'archer':
                        player = Archer(username)
                    elif role.lower() == 'magician':
                        player = Magician(username)
                    else:
                        print("Invalid role. Continuing as Novice.")

                boss = Boss("Monster")  # Reset boss HP for next match

        print(f"{player.getUsername()} has {wins} wins.")

    def player_vs_player_mode(self):
        username1 = input("Enter username for Player 1: ")
        username2 = input("Enter username for Player 2: ")
        player1 = Novice(username1)
        player2 = Novice(username2)
        wins1, wins2 = 0, 0

        while player1.getHp() > 0 and player2.getHp() > 0:
            attacker = random.choice([player1, player2])
            if attacker == player1:
                player1.basicAttack(player2)
            else:
                player2.basicAttack(player1)

            print(f"{player1.getUsername()} HP: {player1.getHp()} | {player2.getUsername()} HP: {player2.getHp()}")

            if player1.getHp() <= 0:
                print(f"{player2.getUsername()} wins!")
                wins2 += 1
                player1.setHp(100)  # Reset HP for next match
                player2.setHp(100)  # Reset HP for next match
            elif player2.getHp() <= 0:
                print(f"{player1.getUsername()} wins!")
                wins1 += 1
                player1.setHp(100)  # Reset HP for next match
                player2.setHp(100)  # Reset HP for next match

        print(f"Final Score: {player1.getUsername()} - {wins1} wins | {player2.getUsername()} - {wins2} wins")


# Start the game
game = Game()
game.start_game()
```
